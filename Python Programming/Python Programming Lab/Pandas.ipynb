## 📌 Section 1: Pandas Overview

**Pandas** is an open-source Python library for **data manipulation and analysis**.  
It provides fast, flexible, and expressive data structures — think of it as Excel/SQL inside Python.

### Why Pandas?
| Feature | Plain Python | Pandas |
|---------|-------------|--------|
| Tabular data | Nested lists | DataFrame |
| Labelled data | Dictionaries | Series / DataFrame |
| CSV / Excel I/O | Manual parsing | One-liner |
| Missing data | Manual handling | Built-in |
| Aggregation | Loops | groupby / agg |
| Merge / Join | Complex code | merge / concat |

### Real-World Use Cases:
- 📊 Data cleaning & wrangling
- 🧮 Financial & statistical analysis
- 🤖 Data preparation for Machine Learning
- 📈 Business intelligence & reporting
- 🗃️ Database-like operations in Python

### 2.1 Series
A **Series** is a one-dimensional labelled array that can hold any data type.

```python
# ── Creating a Series ──

# From a list
s1 = pd.Series([10, 20, 30, 40, 50])
print("Series from list:")
print(s1)
print(f"dtype : {s1.dtype} | index : {list(s1.index)}")
```

```python
# From a list with custom index
s2 = pd.Series([85, 92, 78, 95, 88],
               index=['Maths', 'Physics', 'Chemistry', 'CS', 'English'])
print("Series with custom index:")
print(s2)
print(f"\nAccess by label : s2['Physics'] = {s2['Physics']}")
print(f"Access by pos   : s2[0]          = {s2.iloc[0]}")
```

```python
# From a dictionary
student_cgpa = {'Aarav': 8.9, 'Priya': 9.2, 'Rohan': 7.8, 'Sneha': 9.5, 'Vikram': 8.1}
s3 = pd.Series(student_cgpa)
print("Series from dict:")
print(s3)

# From a scalar (broadcast)
s4 = pd.Series(100, index=['a', 'b', 'c', 'd'])
print("\nSeries from scalar:")
print(s4)
```

### 2.2 DataFrame
A **DataFrame** is a 2-dimensional labelled data structure with columns of potentially different types.  
Think of it as a spreadsheet or SQL table.

```python
# ── Creating a DataFrame ──

# From a dictionary of lists
data = {
    'Name'   : ['Aarav', 'Priya', 'Rohan', 'Sneha', 'Vikram', 'Kavya'],
    'Age'    : [20, 21, 20, 22, 21, 20],
    'Branch' : ['CSE', 'ECE', 'CSE', 'ME', 'CSE', 'ECE'],
    'CGPA'   : [8.9, 9.2, 7.8, 9.5, 8.1, 9.0],
    'Backlogs': [0, 0, 1, 0, 2, 0]
}
df = pd.DataFrame(data)
print("DataFrame from dict:")
print(df)
```

```python
# From a list of dictionaries
records = [
    {'Product': 'Laptop',  'Price': 55000, 'Stock': 30},
    {'Product': 'Phone',   'Price': 20000, 'Stock': 80},
    {'Product': 'Tablet',  'Price': 35000, 'Stock': 50},
    {'Product': 'Monitor', 'Price': 18000, 'Stock': 45},
]
df_prod = pd.DataFrame(records)
print("DataFrame from list of dicts:")
print(df_prod)
```

```python
# From a NumPy array with column names
arr = np.random.randint(40, 100, size=(5, 4))
subjects = ['Maths', 'Physics', 'Chemistry', 'CS']
df_marks = pd.DataFrame(arr, columns=subjects,
                        index=[f'Student_{i+1}' for i in range(5)])
print("DataFrame from NumPy array:")
print(df_marks)
```

## 📌 Section 3: Operations on a Series

### 3.1 head() and tail()

```python
# Create a larger Series for demonstration
np.random.seed(42)
temperatures = pd.Series(
    np.round(np.random.uniform(20, 45, 30), 1),
    index=pd.date_range('2024-01-01', periods=30, freq='D'),
    name='Temperature (°C)'
)

print("Full Series has", len(temperatures), "entries")
print("\n--- head(5) — first 5 rows ---")
print(temperatures.head(5))
print("\n--- tail(5) — last 5 rows ---")
print(temperatures.tail(5))
print("\n--- head(3) ---")
print(temperatures.head(3))
```

### 3.2 Vector Operations on Series

```python
# Basic arithmetic (element-wise, aligned by index)
marks = pd.Series({'Maths': 85, 'Physics': 78, 'Chemistry': 92, 'CS': 95, 'English': 70})
print("Original marks:")
print(marks)

print("\n+10 grace marks  :")
print(marks + 10)

print("\nPercentage /100  :")
print((marks / 100 * 100).round(1))

print("\nSquare of marks  :")
print(marks ** 2)
```

```python
# Statistical operations on a Series
scores = pd.Series([72, 85, 91, 67, 88, 95, 74, 82, 79, 93])
print(f"count  : {scores.count()}")
print(f"sum    : {scores.sum()}")
print(f"mean   : {scores.mean():.2f}")
print(f"median : {scores.median()}")
print(f"std    : {scores.std():.2f}")
print(f"var    : {scores.var():.2f}")
print(f"min    : {scores.min()}")
print(f"max    : {scores.max()}")
print(f"\ndescribe():")
print(scores.describe())
```

```python
# Useful Series methods
s = pd.Series([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])
print("Series    :", list(s))
print("sort_values()    :", list(s.sort_values()))
print("value_counts():")
print(s.value_counts())
print("unique()  :", s.unique())
print("nunique() :", s.nunique())
print("cumsum()  :", list(s.cumsum()))
```

## 📌 Section 4: DataFrame Operations

### 4.1 Creating and Displaying a DataFrame

```python
# Our main working DataFrame for this section
data = {
    'Name'    : ['Aarav', 'Priya', 'Rohan', 'Sneha', 'Vikram', 'Kavya', 'Arjun', 'Meera'],
    'Age'     : [20, 21, 20, 22, 21, 20, 22, 21],
    'Branch'  : ['CSE', 'ECE', 'CSE', 'ME', 'CSE', 'ECE', 'ME', 'CSE'],
    'Maths'   : [85, 78, 92, 70, 88, 95, 65, 90],
    'Physics' : [80, 85, 75, 68, 82, 91, 72, 87],
    'CGPA'    : [8.9, 9.2, 7.8, 8.5, 8.1, 9.5, 7.5, 9.0],
    'Backlogs': [0, 0, 1, 1, 2, 0, 3, 0]
}
df = pd.DataFrame(data)
print("Full DataFrame:")
print(df)
```

```python
# ── Displaying a DataFrame ──
print("--- head(3) ---")
print(df.head(3))

print("\n--- tail(3) ---")
print(df.tail(3))

print("\n--- info() ---")
df.info()
```

```python
print("--- shape ---")
print("Rows, Cols:", df.shape)

print("\n--- columns ---")
print(df.columns.tolist())

print("\n--- dtypes ---")
print(df.dtypes)

print("\n--- describe() ---")
print(df.describe())
```

### 4.2 Iteration over a DataFrame

```python
# Iterate over columns
print("=== Iterating over columns ===")
for col_name in df.columns:
    print(f"Column '{col_name}' — dtype: {df[col_name].dtype}")
```

```python
# iterrows() — iterate over rows as (index, Series)
print("=== iterrows() — first 3 rows ===")
for idx, row in df.head(3).iterrows():
    print(f"Row {idx}: {row['Name']}, Branch={row['Branch']}, CGPA={row['CGPA']}")
```

```python
# itertuples() — faster than iterrows
print("=== itertuples() — first 3 rows ===")
for row in df.head(3).itertuples():
    print(f"Row {row.Index}: {row.Name}, Age={row.Age}, CGPA={row.CGPA}")
```

### 4.3 Selecting Columns

```python
# Select a single column → returns a Series
name_col = df['Name']
print("Single column (Series):")
print(name_col)
print("Type:", type(name_col))
```

```python
# Select multiple columns → returns a DataFrame
subset = df[['Name', 'Branch', 'CGPA']]
print("Multiple columns (DataFrame):")
print(subset)
```

```python
# loc — label-based selection
print("df.loc[0:2, 'Name':'CGPA']:")
print(df.loc[0:2, 'Name':'CGPA'])

# iloc — integer-based selection
print("\ndf.iloc[0:3, 0:4]:")
print(df.iloc[0:3, 0:4])

# Conditional selection
print("\nCSE students with CGPA > 8.5:")
print(df[(df['Branch'] == 'CSE') & (df['CGPA'] > 8.5)])
```

### 4.4 Adding a Column

```python
# Method 1: Direct assignment
df['Total'] = df['Maths'] + df['Physics']
print("Added 'Total' column:")
print(df[['Name', 'Maths', 'Physics', 'Total']].head())
```

```python
# Method 2: Using assign() — returns a new DataFrame
df = df.assign(
    Percentage = lambda x: round(x['Total'] / 200 * 100, 1),
    Pass_Fail  = lambda x: x['Total'].apply(lambda t: 'Pass' if t >= 100 else 'Fail')
)
print("After adding 'Percentage' and 'Pass_Fail':")
print(df[['Name', 'Total', 'Percentage', 'Pass_Fail']])
```

```python
# Method 3: insert() — add at specific position
df.insert(3, 'Chemistry', np.random.randint(60, 100, len(df)))
print("Inserted 'Chemistry' at position 3:")
print(df.head(3))
```

### 4.5 Deleting a Column

```python
# drop() — remove columns or rows
df_temp = df.copy()

# Drop a single column
df_temp = df_temp.drop(columns=['Chemistry'])
print("After dropping 'Chemistry':")
print(df_temp.columns.tolist())

# Drop multiple columns
df_temp2 = df.drop(columns=['Total', 'Percentage', 'Pass_Fail', 'Chemistry'])
print("\nAfter dropping multiple columns:")
print(df_temp2.columns.tolist())

# del keyword (in-place)
df_del = df.copy()
del df_del['Chemistry']
print("\nUsing del — columns:", df_del.columns.tolist())
```

## 📌 Section 5: Binary Operations in a DataFrame

Pandas supports element-wise binary operations between DataFrames, Series, and scalars.

```python
# Base DataFrames for binary operations
df1 = pd.DataFrame({
    'A': [10, 20, 30, 40],
    'B': [5,  15, 25, 35],
    'C': [2,  4,  6,  8]
})
df2 = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 5, 5, 5],
    'C': [2, 2, 2, 2]
})
print("df1:\n", df1)
print("\ndf2:\n", df2)
```

```python
# ── Addition ──
print("df1 + df2 (element-wise):")
print(df1 + df2)

print("\ndf1.add(df2)  (same as +):")
print(df1.add(df2))

print("\ndf1 + 100  (scalar addition / broadcast):")
print(df1 + 100)
```

```python
# ── Subtraction ──
print("df1 - df2:")
print(df1 - df2)

print("\ndf1.sub(df2):")
print(df1.sub(df2))

print("\ndf1.sub(10)  (subtract scalar):")
print(df1.sub(10))
```

```python
# ── Multiplication ──
print("df1 * df2:")
print(df1 * df2)

print("\ndf1.mul(df2):")
print(df1.mul(df2))

print("\ndf1 * 2  (scalar multiply):")
print(df1 * 2)
```

```python
# ── Division ──
print("df1 / df2:")
print(df1 / df2)

print("\ndf1.div(df2):")
print(df1.div(df2))

print("\nFloor division df1 // df2:")
print(df1 // df2)

print("\nModulo df1 % df2:")
print(df1 % df2)
```

## 📌 Section 6: Matching and Broadcasting Operations

Pandas **aligns data by index and column labels** before performing operations.  
This is different from NumPy — pandas is label-aware.

```python
# ── Index Alignment ──
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['b', 'c', 'd', 'e'])

print("s1:", dict(s1))
print("s2:", dict(s2))
print("\ns1 + s2 (aligned on index — NaN where no match):")
print(s1 + s2)

print("\ns1.add(s2, fill_value=0)  (fill missing with 0):")
print(s1.add(s2, fill_value=0))
```

```python
# ── DataFrame Index Alignment ──
dfA = pd.DataFrame({'X': [1, 2, 3], 'Y': [4, 5, 6]},  index=[0, 1, 2])
dfB = pd.DataFrame({'X': [10, 20],  'Y': [30, 40], 'Z': [50, 60]}, index=[1, 2])

print("dfA:\n", dfA)
print("\ndfB:\n", dfB)
print("\ndfA + dfB (NaN where indices/columns don't overlap):")
print(dfA + dfB)

print("\ndfA.add(dfB, fill_value=0):")
print(dfA.add(dfB, fill_value=0))
```

```python
# ── Broadcasting a Series across a DataFrame ──
df_marks = pd.DataFrame({
    'Maths'    : [85, 72, 90],
    'Physics'  : [78, 88, 65],
    'Chemistry': [92, 65, 80]
}, index=['Aarav', 'Priya', 'Rohan'])

grace = pd.Series({'Maths': 5, 'Physics': 10, 'Chemistry': 3})

print("Original Marks:")
print(df_marks)
print("\nGrace marks per subject:", dict(grace))
print("\nAfter adding grace marks (broadcasting along columns):")
print(df_marks.add(grace))

# Broadcasting along rows (axis=0)
bonus = pd.Series([2, 0, 5], index=['Aarav', 'Priya', 'Rohan'])
print("\nBonus per student:", dict(bonus))
print("\nAfter adding bonus (broadcasting along rows):")
print(df_marks.add(bonus, axis=0))
```

## 📌 Section 7: Handling Missing Data and Filling Values

Real-world datasets almost always have **missing values**.  
Pandas represents missing data as `NaN` (Not a Number) or `None`.

```python
# Create a DataFrame with missing values
data_missing = {
    'Name'   : ['Aarav', 'Priya', 'Rohan', 'Sneha', 'Vikram'],
    'Age'    : [20, np.nan, 20, 22, np.nan],
    'CGPA'   : [8.9, 9.2, np.nan, np.nan, 8.1],
    'Salary' : [np.nan, 45000, 38000, np.nan, 42000],
    'City'   : ['Delhi', None, 'Mumbai', 'Pune', None]
}
df_miss = pd.DataFrame(data_missing)
print("DataFrame with missing values:")
print(df_miss)
```

```python
# ── Detecting missing values ──
print("isnull():")
print(df_miss.isnull())

print("\nMissing count per column:")
print(df_miss.isnull().sum())

print("\nMissing % per column:")
print(round(df_miss.isnull().sum() / len(df_miss) * 100, 1))

print("\nAny NaN in each column:", df_miss.isnull().any().to_dict())
print("Any NaN in entire DF  :", df_miss.isnull().any().any())
```

```python
# ── Dropping missing values ──
print("dropna() — drop rows with ANY NaN:")
print(df_miss.dropna())

print("\ndropna(how='all') — drop rows where ALL are NaN:")
print(df_miss.dropna(how='all'))

print("\ndropna(subset=['CGPA']) — drop rows where CGPA is NaN:")
print(df_miss.dropna(subset=['CGPA']))

print("\ndropna(axis=1) — drop columns with any NaN:")
print(df_miss.dropna(axis=1))
```

```python
# ── Filling missing values ──
print("fillna(0) — fill all NaN with 0:")
print(df_miss.fillna(0))
```

```python
# Fill with specific values per column
fill_values = {
    'Age'   : df_miss['Age'].mean(),
    'CGPA'  : df_miss['CGPA'].median(),
    'Salary': df_miss['Salary'].mean(),
    'City'  : 'Unknown'
}
df_filled = df_miss.fillna(fill_values)
print("fillna with column-specific values:")
print(df_filled)
print("\nRemaining NaN count:", df_filled.isnull().sum().sum())
```

```python
# Forward fill and backward fill
df_ff = df_miss.copy()
print("ffill() — propagate last valid value FORWARD:")
print(df_ff.ffill())

print("\nbfill() — propagate next valid value BACKWARD:")
print(df_ff.bfill())

# Interpolate numeric columns
df_interp = df_miss.copy()
df_interp[['Age','CGPA','Salary']] = df_interp[['Age','CGPA','Salary']].interpolate()
print("\ninterpolate() — linear interpolation:")
print(df_interp)
```

## 📌 Section 8: Data Aggregation

**Aggregation** summarises data — finding totals, averages, counts, etc. by groups.

```python
# Build an aggregation-friendly dataset
np.random.seed(10)
df_agg = pd.DataFrame({
    'Name'    : ['Aarav','Priya','Rohan','Sneha','Vikram','Kavya','Arjun','Meera','Dev','Ananya'],
    'Branch'  : ['CSE','ECE','CSE','ME','CSE','ECE','ME','CSE','ECE','ME'],
    'Year'    : [2, 3, 2, 4, 3, 2, 3, 4, 2, 3],
    'CGPA'    : [8.9, 9.2, 7.8, 8.5, 8.1, 9.5, 7.5, 9.0, 8.3, 8.7],
    'Salary'  : [65000, 72000, 55000, 68000, 60000, 80000, 52000, 75000, 62000, 70000],
    'Projects': [3, 5, 2, 4, 3, 6, 2, 4, 3, 5]
})
print(df_agg)
```

```python
# ── groupby() ──
# Group by Branch and compute mean CGPA
branch_avg = df_agg.groupby('Branch')['CGPA'].mean()
print("Average CGPA by Branch:")
print(branch_avg)

print("\nMultiple aggregations by Branch:")
print(df_agg.groupby('Branch').agg({
    'CGPA'    : ['mean', 'min', 'max'],
    'Salary'  : ['mean', 'sum'],
    'Projects': 'sum'
}).round(2))
```

```python
# Group by multiple columns
print("Average CGPA by Branch and Year:")
print(df_agg.groupby(['Branch', 'Year'])['CGPA'].mean().round(2))

# Size / count of groups
print("\nCount of students per Branch:")
print(df_agg.groupby('Branch').size())

print("\nnunique per group:")
print(df_agg.groupby('Branch')['Year'].nunique())
```

```python
# agg() with custom functions
def score_range(x):
    return x.max() - x.min()

print("Custom aggregation (range):")
print(df_agg.groupby('Branch')['CGPA'].agg(['mean', 'std', score_range]).round(3))

# Named aggregation (pandas >= 0.25)
result = df_agg.groupby('Branch').agg(
    Avg_CGPA    = ('CGPA',     'mean'),
    Max_Salary  = ('Salary',   'max'),
    Total_Proj  = ('Projects', 'sum'),
    Headcount   = ('Name',     'count')
).round(2)
print("\nNamed aggregation:")
print(result)
```

```python
# pivot_table — like Excel pivot
pivot = df_agg.pivot_table(
    values  = 'CGPA',
    index   = 'Branch',
    columns = 'Year',
    aggfunc = 'mean'
).round(2)
print("Pivot table — CGPA by Branch × Year:")
print(pivot)
```

## 📌 Section 9: Comparisons & Boolean Reductions

### 9.1 Comparison Operators

```python
df_cmp = pd.DataFrame({
    'A': [10, 20, 30, 40, 50],
    'B': [15, 15, 35, 35, 55],
    'C': [5,  25, 25, 45, 45]
})
print("DataFrame:")
print(df_cmp)

print("\ndf_cmp > 20:")
print(df_cmp > 20)

print("\ndf_cmp == 15:")
print(df_cmp == 15)

print("\ndf_cmp >= 35:")
print(df_cmp >= 35)
```

### 9.2 Boolean Reductions

```python
df_bool = pd.DataFrame({
    'A': [True,  False, True,  True ],
    'B': [True,  True,  False, True ],
    'C': [False, False, False, True ]
})
print("Boolean DataFrame:")
print(df_bool)

print("\nany() per column — at least one True:")
print(df_bool.any())

print("\nall() per column — all True:")
print(df_bool.all())

print("\nany(axis=1) per row:")
print(df_bool.any(axis=1))

print("\nall(axis=1) per row:")
print(df_bool.all(axis=1))

# On numeric DataFrame with condition
print("\nAny value > 30 per column in df_cmp:")
print((df_cmp > 30).any())

print("\nAll values > 0 per column:")
print((df_cmp > 0).all())
```

```python
# Practical: filter with boolean conditions
df_students = pd.DataFrame({
    'Name'    : ['Aarav','Priya','Rohan','Sneha','Vikram','Kavya'],
    'CGPA'    : [8.9, 9.2, 7.8, 8.5, 8.1, 9.5],
    'Backlogs': [0, 0, 1, 1, 0, 0],
    'Branch'  : ['CSE','ECE','CSE','ME','CSE','ECE']
})

# Single condition
print("CGPA > 8.5:")
print(df_students[df_students['CGPA'] > 8.5])

# Multiple conditions (AND)
eligible = df_students[(df_students['CGPA'] >= 8.0) & (df_students['Backlogs'] == 0)]
print("\nPlacement eligible (CGPA>=8.0, no backlogs):")
print(eligible)

# isin()
cse_ece = df_students[df_students['Branch'].isin(['CSE', 'ECE'])]
print("\nCSE or ECE students:")
print(cse_ece)

# between()
print("\nCGPA between 8.0 and 9.0:")
print(df_students[df_students['CGPA'].between(8.0, 9.0)])
```

## 📌 Section 10: Comparing Series

```python
# Element-wise comparison
mid_sem  = pd.Series([72, 85, 68, 90, 78], index=['Aarav','Priya','Rohan','Sneha','Vikram'])
end_sem  = pd.Series([78, 82, 75, 88, 85], index=['Aarav','Priya','Rohan','Sneha','Vikram'])

print("Mid-Sem scores:", dict(mid_sem))
print("End-Sem scores:", dict(end_sem))

print("\nImproved (end > mid)?")
print(end_sem > mid_sem)

print("\nDifference (end - mid):")
print(end_sem - mid_sem)

print("\nnp.maximum (best of both):")
print(np.maximum(mid_sem, end_sem))
```

```python
# equals() — checks exact equality of two Series
s_a = pd.Series([1, 2, 3, 4], index=['a','b','c','d'])
s_b = pd.Series([1, 2, 3, 4], index=['a','b','c','d'])
s_c = pd.Series([1, 2, 3, 5], index=['a','b','c','d'])

print("s_a == s_b (element-wise):", list(s_a == s_b))
print("s_a.equals(s_b)           :", s_a.equals(s_b))
print("s_a.equals(s_c)           :", s_a.equals(s_c))

# compare() — detailed diff (pandas >= 1.1)
diff = s_a.compare(s_c)
print("\ns_a.compare(s_c):")
print(diff if not diff.empty else "No differences")

diff2 = s_a.compare(s_c)
print(diff2)
```

```python
# Rank and correlation
print("Rank of mid_sem scores:")
print(mid_sem.rank(ascending=False))

print("\nCorrelation between mid_sem and end_sem:")
print(f"  corr() = {mid_sem.corr(end_sem):.4f}")

print("\nCumulative sum:")
print(mid_sem.cumsum())

print("\nRolling mean (window=3):")
print(mid_sem.rolling(3).mean())
```

## 📌 Section 11: Combining DataFrames

Pandas provides several ways to combine DataFrames:

| Method | Purpose |
|--------|---------|
| `pd.concat()` | Stack DataFrames vertically or horizontally |
| `pd.merge()` | SQL-style join on a key column |
| `df.join()` | Join on index |
| `df.append()` | Add rows (deprecated; use concat) |

```python
# ── concat() — stacking DataFrames ──
df_batch1 = pd.DataFrame({
    'Name'  : ['Aarav', 'Priya', 'Rohan'],
    'Branch': ['CSE', 'ECE', 'CSE'],
    'CGPA'  : [8.9, 9.2, 7.8]
})
df_batch2 = pd.DataFrame({
    'Name'  : ['Sneha', 'Vikram', 'Kavya'],
    'Branch': ['ME', 'CSE', 'ECE'],
    'CGPA'  : [8.5, 8.1, 9.5]
})

# Vertical stack (rows)
df_all = pd.concat([df_batch1, df_batch2], ignore_index=True)
print("Vertical concat (ignore_index=True):")
print(df_all)
```

```python
# Horizontal stack (columns)
df_marks1 = pd.DataFrame({'Maths'  : [85, 72, 90]}, index=[0,1,2])
df_marks2 = pd.DataFrame({'Physics': [78, 88, 65]}, index=[0,1,2])
df_marks3 = pd.DataFrame({'Chemistry':[92,65,80]},  index=[0,1,2])

df_h = pd.concat([df_marks1, df_marks2, df_marks3], axis=1)
print("Horizontal concat (axis=1):")
print(df_h)
```

```python
# ── merge() — SQL-style joins ──
df_students_info = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105],
    'Name'     : ['Aarav', 'Priya', 'Rohan', 'Sneha', 'Vikram'],
    'Branch'   : ['CSE', 'ECE', 'CSE', 'ME', 'CSE']
})
df_students_marks = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 106],
    'CGPA'     : [8.9, 9.2, 7.8, 8.5, 8.0],
    'Backlogs' : [0, 0, 1, 1, 0]
})
print("Students info:")
print(df_students_info)
print("\nStudents marks:")
print(df_students_marks)
```

```python
# Inner join — only matching keys
inner = pd.merge(df_students_info, df_students_marks, on='StudentID', how='inner')
print("INNER join (only matching StudentIDs):")
print(inner)
```

```python
# Left join — all rows from left
left = pd.merge(df_students_info, df_students_marks, on='StudentID', how='left')
print("LEFT join (all from left, NaN for no-match right):")
print(left)
```

```python
# Right join — all rows from right
right = pd.merge(df_students_info, df_students_marks, on='StudentID', how='right')
print("RIGHT join (all from right, NaN for no-match left):")
print(right)
```

```python
# Outer join — all rows from both
outer = pd.merge(df_students_info, df_students_marks, on='StudentID', how='outer')
print("OUTER join (all rows from both):")
print(outer)
```

```python
# ── join() — join on index ──
df_a = pd.DataFrame({'CGPA': [8.9, 9.2, 7.8]},
                    index=['Aarav','Priya','Rohan'])
df_b = pd.DataFrame({'Salary': [65000, 72000, 55000]},
                    index=['Aarav','Priya','Rohan'])

joined = df_a.join(df_b)
print("join() on index:")
print(joined)
```

```python
# ── to_csv() — export options ──
df_out = df_csv.copy()

# Export without row index
df_out.to_csv('students_noindex.csv', index=False)
print("✅ Exported without index")

# Export only selected columns
df_out[['Name','Branch','CGPA']].to_csv('students_subset.csv', index=False)
print("✅ Exported subset columns")

# Export with custom separator
df_out.to_csv('students_pipe.csv', sep='|', index=False)
print("✅ Exported with pipe separator")

# Read pipe-separated file back
df_pipe = pd.read_csv('students_pipe.csv', sep='|')
print("\nRead pipe-separated file:")
print(df_pipe.head(3))
```

```python
# ── Large file reading with chunks ──
print("Reading in chunks of 3 rows:")
chunk_iter = pd.read_csv('students.csv', chunksize=3)

for i, chunk in enumerate(chunk_iter):
    print(f"\nChunk {i+1} (rows {i*3+1}–{i*3+len(chunk)}):")
    print(chunk[['Name','Branch','CGPA']].to_string(index=False))
```

```python
# ── Full pipeline: Read → Process → Export ──
df_pipeline = pd.read_csv('students.csv')

# Step 1: Add computed columns
df_pipeline['Grade'] = df_pipeline['CGPA'].apply(
    lambda x: 'A+' if x >= 9.5 else ('A' if x >= 9.0 else ('B' if x >= 8.0 else 'C'))
)

# Step 2: Filter
df_pipeline['Eligible'] = (df_pipeline['CGPA'] >= 8.0).map({True: 'Yes', False: 'No'})

# Step 3: Sort
df_pipeline = df_pipeline.sort_values('CGPA', ascending=False).reset_index(drop=True)

print("Processed DataFrame:")
print(df_pipeline[['Name','Branch','CGPA','Grade','Eligible']])

# Step 4: Export final result
df_pipeline.to_csv('students_processed.csv', index=False)
print("\n✅ Final processed data exported to 'students_processed.csv'")
```
