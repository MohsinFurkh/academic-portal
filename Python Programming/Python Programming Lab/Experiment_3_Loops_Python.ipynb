# Experiment 3: Loops (Python Lab)
This notebook contains Python programs demonstrating loop concepts with explanations.

## 1. Factorial of a Number

```
n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial:", fact)
```

## 2. Armstrong Number

```
num = int(input("Enter a number: "))
temp = num
digits = len(str(num))
total = 0
while temp > 0:
    d = temp % 10
    total += d ** digits
    temp //= 10
print("Armstrong" if total == num else "Not Armstrong")
```

## 3. Fibonacci Series

```
n = int(input("Enter number of terms: "))
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

## 4. Prime Number Check

```
num = int(input("Enter a number: "))
flag = num > 1
for i in range(2, num):
    if num % i == 0:
        flag = False
        break
print("Prime" if flag else "Not Prime")
```

## 5. Palindrome Number

```
num = int(input("Enter a number: "))
temp = num
rev = 0
while temp > 0:
    rev = rev * 10 + temp % 10
    temp //= 10
print("Palindrome" if rev == num else "Not Palindrome")
```

## 6. Sum of Digits

```
num = int(input("Enter a number: "))
s = 0
while num > 0:
    s += num % 10
    num //= 10
print("Sum of digits:", s)
```

## 7. Numbers Divisible by 5 or 7

```
count = 0
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1
print("\nCount:", count)
```

## 8. Lowercase to Uppercase

```
text = input("Enter a string: ")
res = ""
for ch in text:
    res += ch.upper() if ch.islower() else ch
print(res)
```

## 9. Multiplication Table

```
num = int(input("Enter a number: "))
for i in range(1, 11):
    print(f"{num} * {i} = {num*i}")
```

## 10. Pattern Printing

```
for i in range(5):
    for j in range(1, 6 - i):
        print(j, end="")
    for k in range(2 * i - 1):
        print("*", end="")
    for j in range(6 - i - 1, 0, -1):
        print(j, end="")
    print()
```

## 11. Sum of Series

```
n = int(input("Enter n: "))
s = 0.0
for i in range(1, n + 1):
    s += 1 / i
print("Sum of series:", s)
```
